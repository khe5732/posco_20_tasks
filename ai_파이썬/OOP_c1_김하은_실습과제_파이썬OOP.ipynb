# 클래스 실습 : 나의 정보 만들기

```python
class Person:
    ## 이름, 나이, 키, 몸무게 받아옴
    ## 초기값은 모두 'none'
    def __init__(self, name='none', age='none', height='none', weight='none'):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
    
    # 나이 1살 더하기
    def age_older(self):
        self.age += 1
    
    # 키 입력한 amout 만큼 더하기
    def change_height(self, amount):
        self.height += amount
    
    # 몸무게 입력한 amozut 만큼 더하기
    def change_weight(self, amount):
        self.weight += amount
    
    # 개인 정보 출력
    def get_info(self):
        print('이름:', self.name)
        print('나이:',self.age)
        print('키:', self.height)
        print('몸무게:', self.weight)
```

```python
# 아무것도 입력하지 않을 때
print('<초기값으로 시작>')
hi = Person()
hi.get_info()
print('-----------------------------')

# 내 정보 입력
print('\n<내 정보 입력 후>')
hi = Person('김하은',25, 155, 45)
hi.get_info()
print('-----------------------------')

# 한 살 더하기
print('\n<한 살 더하기>')
hi.age_older()
hi.get_info()
print('-----------------------------')

# 키에 10 더하기 
print('\n<키에 10cm 더하기>')
hi.change_height(10)
hi.get_info()
print('-----------------------------')

# 몸무게에 5 더하기 
print('\n<몸무게에 5kg 더하기>')
hi.change_weight(5)
hi.get_info()
print('-----------------------------')
```

# 실습 문제 1 : 은행계좌 클래스 만들기

```python
# 실습 문제 1 : 은행계좌 클래스 만들기

class BankAccount:
    # 계좌 클래스의 __init__ 인자는 balance와 name를 가짐
    # 지정 안할 때 초기값 -> balace=0, name='none'
    def __init__(self, balance=0, name='none'):
        self.balance = balance
        self.name = name
    
    # 입금 기능 : balabce(잔액)에 입력한 amount만큼 더하기
    def deposit(self, amount):
        self.balance += amount
        
    # 출금 기능 : balabce(잔액)에 입력한 amount만큼 빼기
    # 단, 잔액보다 amount가 크면 '잔액 부족'이라고 출력 후 출금되지 않음
    def withdraw(self, amount):
        
        if amount > self.balance:
            print('!!! 잔액 부족 !!!')
        else:
            self.balance -= amount
        
    # 계좌 정보 출력    
    def get_info(self):
        print('이름:', self.name)
        print('잔고:',self.balance)
```

```python
c = BankAccount()
c.get_info()
```

```python
# 잔고0, 김하은으로 계좌 생성
a = BankAccount(0,'김하은')
print('<계좌 생성>')
a.get_info()
print('-----------------------------')

# 400원 입금
a.deposit(400)
print('\n<400원 입금>')
a.get_info()
print('-----------------------------')

# 600원 출금 -> '잔액부족' 출력
print('\n<600원 출금>')
a.withdraw(600)
print('-----------------------------')
```

# 실습 문제 2 : 최소 잔액 유지하는 계좌

```python
class MinimumBalanceAccount():
    # 계좌 클래스의 __init__ 인자는 balance와 name를 가짐
    # 지정 안할 때 초기값 -> balace=0, name='none'
    def __init__(self, balance=0, name='none', min_bal=0):
        self.balance = balance
        self.name = name
        self.min_bal = min_bal
        
    # 입금 기능 : balabce(잔액)에 입력한 amount만큼 더하기
    def deposit(self, amount):
        self.balance += amount
        
    # 출금 기능 : balabce(잔액)에 입력한 amount만큼 빼기
    # 단, 잔액보다 amount가 크면 '잔액 부족'이라고 출력 후 출금되지 않음
    def withdraw(self, amount):
        if amount > self.balance:
            print('!!! 잔액 부족 !!!')
        elif (self.balance-amount)<self.min_bal:
            print('최소 잔액을 유지해야 합니다')
        else:
            self.balance -= amount
        
    # 계좌 정보 출력    
    def get_info(self):
        print('이름:', self.name)
        print('잔고:',self.balance)
```

```python
# 잔액 500,이름 'Kim', 최소 잔액 1000인 계좌 생성
acc = MinimumBalanceAccount(500, 'Kim', 1000)
acc.get_info()
print('-----------------------------')

# 1000원 입금 -> 1500원
print('\n<1000원 입금>')
acc.deposit(1000)
acc.get_info()
print('-----------------------------')

# 600원 출금 -> 잔액 900원으로 최소 잔액보다 적어서 에러
print('\n<600원 출금>')
acc.withdraw(600)
acc.get_info()
print('-----------------------------')
```

# 과제 : Python의 기본 자료구조인 Set를 모방한 클래스 Set 구현

```python
class Set:
    # self.lis에서 값이 모두 바뀌어서 이를 그대로 넣은 self.Set에서 출력될 예정
    def __init__(self, lis=[]):
        self.lis = []
        # 중복 제거
        for item in lis:
            if item not in self.lis:
                self.lis.append(item)   
        # 만든 리스트 요소들 그대로 넣기
        self.Set = {item for item in self.lis}
        
    def add(self, elem):
        # 리스트에 append하기
        self.lis.append(elem)
        # 리스트 요소 그대로 넣기
        self.Set = {item for item in self.lis}
        return self
        
    def discard(self, elem):
        # 리스트에 해당 요소가 있으면 삭제
        if elem in self.lis:
            self.lis.remove(elem)
            self.Set = {item for item in self.lis}
        else:
            pass
        
    def clear(self):
        # 리스트 요소 모두 삭제
        self.lis.clear()
        # 집합도 모두 삭제
        self.Set = {}
        
    def __len__(self):
        # 집합 리스트화 -> len함수 사용
        length = list(self.Set)
        return len(length)
    
    def __str__(self):
        # 프린트 할때 집합 출력
        a = str(self.Set)
        return a
    
    def __contains__(self,elem):
        # 포함하면 True, 아니면 False
        return (elem in self.Set)
    
    def __le__(self, other):
        # self <= other : self가 other의 부분집합이면 True, 아니면 False
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        result_list = [item for item in self_list if item not in other_list]
        if len(result_list) == 0:
            answer = 'True'
        else:
            answer = 'False'
        return answer
            
    
    def __ge__(self, other):
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        result_list = [item for item in other_list if item not in self_list]
        if len(result_list) == 0:
            answer = 'True'
        else:
            answer = 'False'
        return answer
            
    def __or__(self, other):
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        result_list = self_list+[item for item in other_list if item not in self_list]
        result = {item for item in result_list}
        return result
    
    def __and__(self, other):
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        result_list = [item for item in self_list if item in other_list]
        result = {item for item in result_list}
        return result
        
    def __sub__(self, other):
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        result_list = [item for item in self_list if item not in other_list]
        result = {item for item in result_list}
        return result
    
    def __ior__(self, other):
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        self.lis = self_list+[item for item in other_list if item not in self_list]
        self.Set = {item for item in self.lis}
        return self
    
    def __iand__(self, other):
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        self.lis = [item for item in self_list if item in other_list]
        self.Set = {item for item in self.lis}
        return self
    
    def __isub__(self, other):
        self_list = list(self.lis)
        if type(other) == set:
            other_list = list(other)
        else:
            other_list = list(other.Set)
        self.lis = [item for item in self_list if item not in other_list]
        self.Set = {item for item in self.lis}
        return self
```
