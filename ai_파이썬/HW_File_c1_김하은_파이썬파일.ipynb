```python
# 1. 파일에 있는 각각의 단어 수 구하기
```

```python
test = open('test.txt','w')

for word in ['first', 'second', 'third']:
    data = '{} line\n'.format(word)
    test.write(data)
    
test.close()
```

```python
test = open('test.txt','r')

data  = test.readlines()
# str 나눠서 단어 가진 list 만들기
lis = []
for words in data:
    word_lis = words.split()
    for word in word_lis:
        lis.append(word)

# lis안의 요소가 key인 딕셔너리 만들기
result = {}
for word in lis:
    if word in result:
        # 이미 딕셔너리에 있으면 1 더하기
        result[word] += 1
    else:
        # 처음이면 1
        result[word] = 1
        

for word in result:
    print(word, result[word])

test.close()
```

```python
# 2. 파일명을 입력 받음 -> 한 줄씩 읽어 파일의 내용 모두 대문자로 출력
## 단, 파일이 없는 경우 파일이 존재하지 않음 출력
```

```python
import os
file = input('Enter a file name:')

#  파일 있는 지 확인
a = os.path.exists(file)
# 없음
if a == False:
    print('No file\n')
# 있음
else:
    print('file exists?:',a, '\n')
    
test = open(file,'r')
data  = test.readlines()

# 한줄 씩 가져와서 대문자화
for words in data:
    print(words.upper(), end='')
test.close()
```

```python
# 실습 4. score.txt를 읽어서 학생들의 성적처리 -> report.txt로 출력
```

```python
# score 데이터 생성
import random

score = open('score.txt','w')

# 학생 수 : 랜덤하게
stu_num = random.randrange(10, 101)

# 점수도 랜덤하게 (단, 모두 F가 될 것을 고려하여 조금 높게 잡음)
for i in range(stu_num):
    if i < 10:
        data = '20220{0} {1} {2}\n'.format(str(i), str(random.randrange(50, 101)),str(random.randrange(50, 101)))
    else:
        data = '2022{0} {1} {2}\n'.format(str(i), str(random.randrange(50, 101)),str(random.randrange(50, 101)))
    score.write(data)

score.close()
```

```python
score = open('score.txt','r')
report = open('report.txt','w')

data  = score.readlines()
for words in data:
    word_lis = words.split()
    
    # 총점 구하기
    total_score = round(int(word_lis[1])*0.4 + int(word_lis[2])*0.6,1)
    
    # 성적 구하기
    if total_score >= 90:
        grade = 'A'
    elif total_score >= 80:
        grade = 'B'
    elif total_score >= 70:
        grade = 'C'
    elif total_score >= 60:
        grade = 'D'
    else:
        grade = 'F'
        
    # 총점, 성적 합쳐서 report 데이터 만들기
    data = '{0} {1} {2} {3}({4})\n'.format(word_lis[0], word_lis[1], word_lis[2], str(total_score),grade)
    report.write(data)
# 
score.close() 
report.close()
```

```python
# report 데이터 확인
report = open('report.txt','r')

data  = report.readlines()
for words in data:
    print(words)
    
score.close() 
report.close()
```
